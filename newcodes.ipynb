```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = [[1],[2],[3],[4]]
y = [2,4,6,8]
model.fit(X,y)
user_value = float(input("Enter a number to predict: "))
prediction = model.predict([[user_value]])
print("Prediction is:", prediction[0])
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y, color="blue", label='Data Points')
plt.plot(X, model.predict(X), color='orange', label='Regression Line')
plt.legend(loc='center left',bbox_to_anchor=(1,0.5))
plt.title("Linear Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Corrected arrays
weights = np.array([1,2,3,4,5]).reshape(-1,1)
prices = np.array([30,60,90,120,150])

plt.scatter(weights, prices, color="blue", label='Data Points')
plt.plot(weights, prices, color='orange', label='Regression Line')
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.title("Linear Regression Line")
plt.xlabel("Weights")
plt.ylabel("Prices")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # features must be 2D
y = np.array([2, 4, 5, 4, 5])                # target values

# Create and train the model
model = LinearRegression()
model.fit(X, y)  # <- THIS IS REQUIRED

# Plot
plt.scatter(X, y, color="blue", label='Data Points')
plt.plot(X, model.predict(X), color='orange', label='Regression Line')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Linear Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()

#EXTRAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAA
```
